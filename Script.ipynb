분석1. 지역에 따라 평균 비디오게임 판매량에 차이가 있는가?

1. page  
4개 지역에 따라 판매된 비디오게임의 양은 다음과 같습니다.🚀  
비디오게임 판매량이 가장 많은 지역은 북미지역입니다.🚀  
일원분산분석결과 지역별 비디오게임의 평균 판매량의 차이는 유의미한 것으로 나타났고,🚀  
사후분석 결과 모든 지역의 차이가 유의미하였습니다.🚀   

1. page  
그래서 지역간 비디오게임의 평균 판매량은 다릅다고 결론을 내릴 수 있고,  
북미지역, PAL 지역, 일본지역, 기타지역의 순서로 비디오게임의 평균 판매량이 높습니다.🚀  

1. page  
시간에 따라서 각 지역별 비디오게임 판매량이 어떻게 변화하였는지를 아래의 그래프로 살펴 볼 수 있습니다.🚀  
역시 파란색으로 표시된 북미지역의 판매량이 높은 것을 볼 수 있습니다.  
그리고 시간이 지남에 따라서 북미지역과 PAL 지역은 판매량이 변화하였지만, 초록색으로 표시된 일본 지역은 판매량에 변화가 거의 없습니다.🚀  
또한 2011년 이후에는 북미지역과 PAL지역의 판매량의 차이가 없어보이고, 일본과 기타지역의 판매량에도 차이가 없어보입니다.🚀  

1. page  
그래서 2011년 이후로 한정하여 지역별 평균판매량의 차이를 분석하면🚀 다음과 같이,  
북미지역과 PAL지역은 평균판매량의 차이가 없어지고, 일본과 기타지역의 평균판매량의 차이도 없어집니다.🚀  
그래서 2011년 이후에는 북미지역과 PAL지역의 평균 비디오게임 판매량이 비슷하게 높고, 일본지역과 기타지역은 비슷하게 낮다고 할 수 있습니다.🚀

1. page  
4개 지역의 비디오게임 장르별 시장점유율을 살펴보면 다음과 같습니다.🚀  
북미지역, PAL지역, 기타지역은 Action 장르의 시장점유율이 가장 높고, 일본지역은 RPG 장르가 가장 높습니다.🚀   

1. page  
지역에 따라서 장르별 평균 판매량에 유의미한 차이가 있는지 알아보기 위하여 이원분산분석을 실시한 결과 다음과 같이 지역과 장르의 상호작용항이 유의미하게 나타났습니다.🚀   
그래서 지역에 따라 장르별 평균 판매량에 차이가 있다고 볼 수 있습니다.🚀

그래프를 통하여 더 자세히 살펴보면,🚀  
북미지역과 PAL지역은 비슷한 분포를 보이지만,🚀

PAL 지역은 북미지역과 비슷한 선호를 보입니다.🚀   
사후검정에서도 PAL 지역에서 평균적으로 판매량이 높은 게임은 엑션, 어드벤쳐, RPG, 그리고 스포츠🚀 라고나타났습니다.🚀

사후 검정에서도  RPG 가 가장 인기있는 장르로 나타났고, 엑션게임의 평균판매량이🚀 가장 적은 것으로 나타났습니다.🚀   

1. page

하지만 사후검정 결과 기타지역에서는 특별하게 인기있는 장르가 없다고 나타났습니다.🚀  
그래서 기타지역에서는 여러 장르가 평균적으로 비슷하게 판매된다고 볼 수 있습니다.🚀

  - 닌텐도는 전년동기대비 매출과 영업이익이 4.7%, 15.1% 감소
  - 텐센트(Tencent)의 2022년 2분기 매출은 3%, 영업이익은 43% 하락
  - 중국 게임산업의 이용자 수가 2008년 집계 이후 처음으로 감소

  - LTE를 지원하게 된 최초의 아이폰
