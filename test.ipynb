```python
# ì‚¬ì „í™•ë¥  : ì•”ì— ê±¸ë¦´ í™•ë¥  1%
# P(C) = 0.01  
prio = 0.01 

# ë°ì´í„°
# ì•” ê²€ì‚¬ê°€ ì •í™•íˆ ì•”ì„ ì–‘ì„±ì´ë¼ê³  ì§„ë‹¨í•  í™•ë¥ ì€ 90%
# P(Pr | C) = 0.9
pos_prob = 0.9

# ì•”ì´ ì•„ë‹Œë° ì–‘ì„±ì´ë¼ê³  ì§„ë‹¨í•  í™•ë¥ ì€ 10%
# P(Pr | ^C) = 0.1
nag_prob = 0.1

# ì‚¬í›„í™•ë¥ 
# P(C | Pr) = P(Pr â‹‚ C) / P(Pr)
#          = (P(Pr | C) * P(C)) / P(Pr)
#          = (P(Pr | C) * P(C)) / P(Pr â‹‚ C) + P(Pr â‹‚ ^C)
#          = (P(Pr | C) * P(C)) / P(Pr | C) * P(C) + P(Pr | ^C) * (1 - P(C))

post_prob = (pos_prob * prio) / (pos_prob * prio + nag_prob * (1 - prio))
print('%.3f'%(post_prob))
```
