최근 실시한 조사에 따르면 국가 A의 실업률이 4%라고 합니다. 
국가 A의 한 도시에서, 도시의 시장은 이 실업률이 자신의 도시에게도 똑같은지 살펴보려 합니다. 
시장은 표본을 모아서 실업률이 4%보다 유의미하게 높은지 알아보고자 합니다.

이 때, p는 도시의 실업률입니다. 
이 실험의 가설은 아래와 같습니다.

H_0 : p = 0.04
H_a : p ≠ 0.04

📌 귀무가설과 대립가설을
H_0 : p =< 0.04
H_a : p > 0.04
으로 설정하는 것이 더 적절하지 않을까?

# 12

```python
# 사전확률 : 암에 걸릴 확률 1%
# P(C) = 0.01  
prio = 0.01 

# 데이터
# 암 검사가 정확히 암을 양성이라고 진단할 확률은 90%
# P(Pr | C) = 0.9
pos_prob = 0.9

# 암이 아닌데 양성이라고 진단할 확률은 10%
# P(Pr | ^C) = 0.1
nag_prob = 0.1

# 사후확률
# P(C | Pr) = P(Pr ⋂ C) / P(Pr)
#          = (P(Pr | C) * P(C)) / P(Pr)
#          = (P(Pr | C) * P(C)) / P(Pr ⋂ C) + P(Pr ⋂ ^C)
#          = (P(Pr | C) * P(C)) / P(Pr | C) * P(C) + P(Pr | ^C) * (1 - P(C))

post_prob = (pos_prob * prio) / (pos_prob * prio + nag_prob * (1 - prio))
print('%.3f'%(post_prob))
```

뺑소니 사건이 발생했습니다. 목격자의 진술에 의하면, 용의차량은 빨간색 차량이라고 합니다. 
해당 지역의 차량은 빨강과 파랑 두 가지로, 전체 차량 중 빨간색 차량의 비율은 20%, 파란색 차량의 비율은 80%를 차지합니다. 

목격자 증언을 검증하기 위해, 사건현장과 비슷한 환경을 구성하여 모의실험을 진행했습니다. 
이때, 실제 빨간색 차량을 빨강이라고 정확히 증언한 비율은 60%로 밝혀졌습니다. 
경찰 측에서는 목격자가 차량의 색을 정확히 증언했을 때, 그 차량이 빨간색 차량일 확률을 알고자 합니다.

(용의차량이 실제로 빨간색 차량일 사건을 R, 용의차량이 실제로 파란색 차량일 사건을 B, 
  목격자가 차량의 색을 정확히 증언하는 사건을 A 라고 정의합니다.)

📌 문제에서...
"용의차량이 실제로 빨간색 차량일 사건을 R"라는 표현이 사전확률을 의미하는지, 사후확률을 의미하는지

```python
# 사전확률 : 전체 차량 중 빨간색 차량의 비율은 20%, 파란색 차량의 비율은 80%
# P(C) = 0.2  
prio = 0.2 

# 데이터
# 암 검사가 정확히 암을 양성이라고 진단할 확률은 90%
# P(Pr | C) = 0.9
pos_prob = 0.9

# 암이 아닌데 양성이라고 진단할 확률은 10%
# P(Pr | ^C) = 0.1
nag_prob = 0.1

# 사후확률
# P(C | Pr) = P(Pr ⋂ C) / P(Pr)
#          = (P(Pr | C) * P(C)) / P(Pr)
#          = (P(Pr | C) * P(C)) / P(Pr ⋂ C) + P(Pr ⋂ ^C)
#          = (P(Pr | C) * P(C)) / P(Pr | C) * P(C) + P(Pr | ^C) * (1 - P(C))

post_prob = (pos_prob * prio) / (pos_prob * prio + nag_prob * (1 - prio))
print('%.3f'%(post_prob))
```
